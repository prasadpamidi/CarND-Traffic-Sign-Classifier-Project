### Perform random rotation


````python
import cv2 as cv

def rotate(image):
    '''
        Rotates the image with a random value
    '''
    c_x,c_y = int(image.shape[0]/2), int(image.shape[1]/2)
    ang = 30.0*np.random.rand()-15
    Mat = cv.getRotationMatrix2D((c_x, c_y), ang, 1.0)
    return cv.warpAffine(img, Mat, image.shape[:2])

%matplotlib inline

img = random.choice(X_train)
dst = rotate(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Perform random translation


````python
def translate(image):
    '''
        Translates the image with a random value
    '''
    rows, cols = image.shape[0], image.shape[1]

    translation_x = 10*np.random.uniform()-5
    translation_y = 10*np.random.uniform()-5
    M = np.float32([[1,0,translation_x], [0,1,translation_y]])
    return cv.warpAffine(img, M, (cols, rows))

%matplotlib inline

img = random.choice(X_train)
dst = translate(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Perform random warping to images


````python
def warping_transform(image):
    '''
        Affine tranform the image
    '''
    rows,cols,_ = image.shape

    rnd_x = np.random.rand(3) - 0.5
    rnd_x *= cols * 0.065
    rnd_y = np.random.rand(3) - 0.5
    rnd_y *= rows * 0.065

    x1 = cols/4
    x2 = 3*cols/4
    y1 = rows/4
    y2 = 3*rows/4

    pts1 = np.float32([[y1,x1],[y2,x1],[y1,x2]])
    pts2 = np.float32([[y1+rnd_y[0],x1+rnd_x[0]],[y2+rnd_y[1],x1+rnd_x[1]],[y1+rnd_y[2],x2+rnd_x[2]]])

    M = cv.getAffineTransform(pts1,pts2)
    
    return cv.warpAffine(img,M,(cols,rows))

%matplotlib inline

img = random.choice(X_train)
dst = warping_transform(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Perform random scaling on images


````python
def scale(image):
    '''
        Scales the image
    '''
    rows, cols = image.shape[0], image.shape[1]

    px = np.random.randint(-2,2)
    pts1 = np.float32([[px,px],[rows-px,px],[px,cols-px],[rows-px,cols-px]])
    pts2 = np.float32([[0,0],[rows,0],[0,cols],[rows,cols]])
    M = cv.getPerspectiveTransform(pts1,pts2)
    dst = cv.warpPerspective(image,M,(rows,cols))
    
    return dst

%matplotlib inline

img = random.choice(X_train)
dst = scale(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Adjust brightness

````python
def brightness(image):
    dst = cv.cvtColor(image,cv.COLOR_RGB2HSV)
    random_bright = .25+np.random.uniform()
    dst[:,:,2] = dst[:,:,2]*random_bright
    dst = cv.cvtColor(dst,cv.COLOR_HSV2RGB)
    return dst

%matplotlib inline

img = random.choice(X_train)
dst = brightness(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Perform augmentation on an image


````python
def augment(image):
    '''
        This method augments the given image by applying rotation, translation, warping and scaling
    '''
    image=brightness(image)
    image=rotate(image)
    image=translate(image)
    return warping_transform(image)

%matplotlib inline

img = random.choice(X_train)
dst = augment(img)

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

### Data Augmentation

````python
'''
 I have picked a random set of images for each class that has sample count less than 750 and performed augmentation. 
 Later I added the augmented dataset into train set to even the data distribution across classes
'''

print('Total training set before augmentation {}'.format(len(X_train)))

for class_n in range(n_classes):
    class_indices = np.where(y_train == class_n)
    n_samples = len(class_indices[0])
    if n_samples < 750:
        for i in range(750):
            img = X_train[class_indices[0][i % n_samples]]
            img = augment(img)
            X_train = np.concatenate((X_train, [img]), axis=0)
            y_train = np.concatenate((y_train, [class_n]), axis=0)
            
print("Data Augmentation Complete")
print('Total training set after augmentation {}'.format(len(X_train)))
````

### Proprocessing (shuffling, grayscale and normalize)

````python
from sklearn.utils import shuffle
from numpy import newaxis

##shuffling data to include some randomness and spread around the augmented data
X_train, y_train = shuffle(X_train, y_train)
    
def grayscale(image):
    '''
        Turns an rgb/bgr image to gray scale
    '''
    return cv.cvtColor(image, cv.COLOR_BGR2GRAY) 

def normalize(image):
    '''
        Normalized an image to gray scale
    '''
    return (image.astype(np.float32) - 128)/128  

def preprocess(dataset):
    '''
        This method go over each image in the dataset and applies grayscale and normalization to them
    '''
    processed_output = []
    for image in dataset:
        processed_output.append(normalize(grayscale(image)))
    return processed_output
        
#Preprocess all the images in train, validation and test datasets. Also, add third dimension with 1 channel(gray scale only)

X_train = np.array(preprocess(X_train))[..., newaxis]
X_valid = np.array(preprocess(X_valid))[..., newaxis]
X_test = np.array(preprocess(X_test))[..., newaxis]

print('Images pre-processing complete')
````

### Model Architecture

````python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils

#Generate one hot encoding values for the classes in the traning, validation and test batch data
y_train = keras.utils.to_categorical(y_train, n_classes)
y_valid = keras.utils.to_categorical(y_valid, n_classes)
y_test = keras.utils.to_categorical(y_test, n_classes)

# Create a forward path neural network layer
model = Sequential()

# Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
model.add(Convolution2D(6, kernel_size=(5, 5), activation='relu', input_shape=(32, 32, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Layer 2: Convolutional. Output = 10x10x16.
model.add(Convolution2D(16, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 6)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten. Input = 5x5x16. Output = 400.
model.add(Flatten())

# Layer 3: Fully Connected. Input = 400. Output = 120.
model.add(Dense(120, activation='relu'))

# Apply dropout
model.add(Dropout(0.65))

# Layer 4: Fully Connected. Input = 120. Output = 84.
model.add(Dense(84, activation='relu'))

# Apply dropout
model.add(Dropout(0.65))

# SOLUTION: Layer 5: Fully Connected. Input = 84. Output = 43.
model.add(Dense(n_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
````

### Train and Validate Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
EPOCHS = 50
BATCH_SIZE = 128

model.fit(X_train, y_train,
          batch_size=BATCH_SIZE,
          epochs=EPOCHS,
          verbose=1,
          validation_data=(X_valid, y_valid))
````

### Model Evaluation

````python
def evaluate(X_data, y_data):
    score = model.evaluate(X_data, y_data, verbose=0)
    return score
````

### Test Model

````python
test_accuracy = evaluate(X_test, y_test)
print("Test Accuracy = {:.3f}".format(test_accuracy[1]))
````
